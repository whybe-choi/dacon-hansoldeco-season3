```python
import os
import re
import pandas as pd
```

# Data Load

```python
data_path = "../data"
test_path = "../data/test.csv"  # org test data

processed_filename = "test_preprocessed.csv"  # 전처리 파일 저장
if os.path.exists(f"{data_path}/{processed_filename}"):
    raise FileExistsError(f"'{processed_filename}' already exists.")

test = pd.read_csv(test_path, encoding="utf-8-sig")
```

# Data Preprocessing

- `공사종류` `공종` `사고객체` `장소` `부위` `인적사고`

    - `인적사고` 는 `(` 뒤에 내용 제거 (괄호 내용 제거)

    - 그 외 컬럼은 각 기호를 기준으로 분리

- test 에는 train 처리 기준을 적용

- 결측치(`np.nan`)는 `없음` 으로 채움

## 공사종류

```python
temp_list = [item.strip() for item in test["공사종류"].fillna("없음").tolist()]

slash_bm_1 = []
slash_bm_2 = []
slash_bm_3 = []

for item in temp_list:
    slash_count = item.count(" / ")
    # 슬래시(" / ")가 1개인 경우
    if slash_count == 1:
        slash_split = item.split(" / ")
        slash_bm_1.append(slash_split[0].strip())
        slash_bm_2.append(slash_split[1].strip())
    # 슬래시(" / ")가 2개인 경우
    elif slash_count == 2:
        slash_split = item.split(" / ")
        slash_bm_1.append(slash_split[0].strip())
        slash_bm_2.append(slash_split[1].strip())
        slash_bm_3.append(slash_split[2].strip())

slash_bm_1 = sorted(list(set(slash_bm_1)))
slash_bm_2 = sorted(list(set(slash_bm_2)))
slash_bm_3 = sorted(list(set(slash_bm_3)))


slash_final_1 = []
slash_final_2 = []
slash_final_3 = []

for item in temp_list:
    slash_split = item.split(" / ")

    # slash_bm_1 기준
    if len(slash_split) >= 1:  # 공사종류1 에 해당하는게 있는지
        found = False
        for x in slash_bm_1:
            if x == slash_split[0].strip():
                found = True
                break
        if found:
            slash_final_1.append(slash_split[0].strip())
        else:
            slash_final_1.append("없음")
    else:
        slash_final_1.append("없음")

    # slash_bm_2 기준
    if len(slash_split) >= 2:
        found = False
        for y in slash_bm_2:
            if y == slash_split[1].strip():
                found = True
                break
        if found:
            slash_final_2.append(slash_split[1].strip())
        else:
            slash_final_2.append("없음")
    else:
        slash_final_2.append("없음")

    # slash_bm_3 기준
    if len(slash_split) >= 3:
        found = False
        for z in slash_bm_3:
            if z == slash_split[2].strip():
                found = True
                break
        if found:
            slash_final_3.append(slash_split[2].strip())
        else:
            slash_final_3.append("없음")
    else:
        slash_final_3.append("없음")


a = pd.DataFrame(
    {"공사종류1": slash_final_1, "공사종류2": slash_final_2, "공사종류3": slash_final_3}
)
```

## 공종

```python
temp_list = [item.strip() for item in test["공종"].fillna("없음").tolist()]

slash_bm_1 = []
slash_bm_2 = []

for item in temp_list:
    slash_count = item.count(">")
    # 슬래시(">")가 0개인 경우
    if slash_count == 0:
        slash_bm_1.append(item.strip())
        slash_bm_2.append("없음")
    # 슬래시(">")가 1개인 경우
    elif slash_count == 1:
        slash_split = item.split(">")
        slash_bm_1.append(slash_split[0].strip())
        slash_bm_2.append(slash_split[1].strip())

slash_bm_1 = sorted(list(set(slash_bm_1)))
slash_bm_2 = sorted(list(set(slash_bm_2)))


slash_final_1 = []
slash_final_2 = []

for item in temp_list:
    slash_split = item.split(">")

    # slash_bm_1 기준
    if len(slash_split) >= 1:
        found = False
        for x in slash_bm_1:
            if x == slash_split[0].strip():
                found = True
                break
        if found:
            slash_final_1.append(slash_split[0].strip())
        else:
            slash_final_1.append("없음")
    else:
        slash_final_1.append("없음")

    # slash_bm_2 기준
    if len(slash_split) >= 2:
        found = False
        for y in slash_bm_2:
            if y == slash_split[1].strip():
                found = True
                break
        if found:
            slash_final_2.append(slash_split[1].strip())
        else:
            slash_final_2.append("없음")
    else:
        slash_final_2.append("없음")


b = pd.DataFrame({"공종1": slash_final_1, "공종2": slash_final_2})
```

## 사고객체

```python
temp_list = [item.strip() for item in test["사고객체"].fillna("없음").tolist()]

slash_bm_1 = []
slash_bm_2 = []

for item in temp_list:
    slash_count = item.count(">")
    # 슬래시(">")가 0개인 경우
    if slash_count == 0:
        slash_bm_1.append(item.strip())
        slash_bm_2.append("없음")
    # 슬래시(">")가 1개인 경우
    elif slash_count == 1:
        slash_split = item.split(">")
        slash_bm_1.append(slash_split[0].strip())
        slash_bm_2.append(slash_split[1].strip())

slash_bm_1 = sorted(list(set(slash_bm_1)))
slash_bm_2 = sorted(list(set(slash_bm_2)))


slash_final_1 = []
slash_final_2 = []

for item in temp_list:
    slash_split = item.split(">")

    # slash_bm_1 기준
    if len(slash_split) >= 1:
        found = False
        for x in slash_bm_1:
            if x == slash_split[0].strip():
                found = True
                break
        if found:
            slash_final_1.append(slash_split[0].strip())
        else:
            slash_final_1.append("없음")
    else:
        slash_final_1.append("없음")

    # slash_bm_2 기준
    if len(slash_split) >= 2:
        found = False
        for y in slash_bm_2:
            if y == slash_split[1].strip():
                found = True
                break
        if found:
            slash_final_2.append(slash_split[1].strip())
        else:
            slash_final_2.append("없음")
    else:
        slash_final_2.append("없음")


c = pd.DataFrame({"사고객체1": slash_final_1, "사고객체2": slash_final_2})
```

## 장소

```python
temp_list = [item.strip() for item in test["장소"].fillna("없음").tolist()]

slash_final_1 = []
slash_final_2 = []

for item in temp_list:
    slash_count = item.count(" / ")

    # 슬래시가 1개인 경우 (텍스트가 왼쪽에 있는 경우)
    if item.endswith("/"):
        left_text = item.strip("/").strip()
        if not left_text == "":
            slash_final_1.append(left_text)
        else:
            slash_final_1.append("없음")
        slash_final_2.append("없음")

    # 슬래시가 1개인 경우 (텍스트가 오른쪽에 있는 경우)
    elif item.startswith("/"):
        right_text = item.strip("/").strip()
        if not right_text == "":
            slash_final_2.append(right_text)
        else:
            slash_final_2.append("없음")
        slash_final_1.append("없음")

    # 슬래시가 1개인 경우
    elif slash_count == 1:
        pattern = re.compile(r"\S")
        left, right = item.split(" / ")
        left_text = left.strip()
        right_text = right.strip()
        if pattern.search(left_text) and pattern.search(right_text):
            if not left_text == "":
                slash_final_1.append(left_text)
            else:
                slash_final_1.append("없음")
            if not right_text == "":
                slash_final_2.append(right_text)
            else:
                slash_final_2.append("없음")

    # 슬래시가 2개인 경우
    elif slash_count == 2:
        pattern = re.compile(r"\S")
        text1, text2, text3 = item.split(" / ")
        if pattern.search(text1) and pattern.search(text2) and pattern.search(text3):
            texts = item.split(" / ", 1)
            left_text = texts[0].strip()
            right_text = texts[1].strip()
            if not left_text == "":
                slash_final_1.append(left_text)
            else:
                slash_final_1.append("없음")
            if not right_text == "":
                slash_final_2.append(right_text)
            else:
                slash_final_2.append("없음")

    # 슬래시만 있는 경우
    elif item.count("/") == 1:
        pattern = re.compile(r"^/$")
        if pattern.fullmatch(item):
            slash_final_1.append("없음")
            slash_final_2.append("없음")


d = pd.DataFrame({"장소1": slash_final_1, "장소2": slash_final_2})
```

## 부위

```python
temp_list = [item.strip() for item in test["부위"].fillna("없음").tolist()]

slash_final_1 = []
slash_final_2 = []

for item in temp_list:
    slash_count = item.count(" / ")

    # 슬래시가 1개인 경우 (텍스트가 왼쪽에 있는 경우)
    if item.endswith("/"):
        left_text = item.strip("/").strip()
        if not left_text == "":
            slash_final_1.append(left_text)
        else:
            slash_final_1.append("없음")
        slash_final_2.append("없음")

    # 슬래시가 1개인 경우 (텍스트가 오른쪽에 있는 경우)
    elif item.startswith("/"):
        right_text = item.strip("/").strip()
        if not right_text == "":
            slash_final_2.append(right_text)
        else:
            slash_final_2.append("없음")
        slash_final_1.append("없음")

    # 슬래시가 1개인 경우
    elif slash_count == 1:
        pattern = re.compile(r"\S")
        left, right = item.split(" / ")
        left_text = left.strip()
        right_text = right.strip()
        if pattern.search(left_text) and pattern.search(right_text):
            if not left_text == "":
                slash_final_1.append(left_text)
            else:
                slash_final_1.append("없음")
            if not right_text == "":
                slash_final_2.append(right_text)
            else:
                slash_final_2.append("없음")

    # 슬래시가 2개인 경우
    elif slash_count == 2:
        pattern = re.compile(r"\S")
        text1, text2, text3 = item.split(" / ")
        if pattern.search(text1) and pattern.search(text2) and pattern.search(text3):
            texts = item.split(" / ", 1)
            left_text = texts[0].strip()
            right_text = texts[1].strip()
            if not left_text == "":
                slash_final_1.append(left_text)
            else:
                slash_final_1.append("없음")
            if not right_text == "":
                slash_final_2.append(right_text)
            else:
                slash_final_2.append("없음")

    # 슬래시가 3개 이상인 경우
    elif item.count("/") >= 3:
        if item.count(" / ") != 1:
            texts = item.split(" / ", 1)
            left_text = texts[0].strip()
            right_text = texts[1].strip()
            if not left_text == "":
                slash_final_1.append(left_text)
            else:
                slash_final_1.append("없음")
            if not right_text == "":
                slash_final_2.append(right_text)
            else:
                slash_final_2.append("없음")

    # 슬래시만 있는 경우
    elif item.count("/") == 1:
        pattern = re.compile(r"^/$")
        if pattern.fullmatch(item):
            slash_final_1.append("없음")
            slash_final_2.append("없음")


e = pd.DataFrame({"부위1": slash_final_1, "부위2": slash_final_2})
```

## 인적사고

```python
temp_list = test["인적사고"].fillna("없음").unique().tolist()
temp_items = {}

for i in temp_list:
    temp_items[i] = i.split("(")[0]

f = pd.DataFrame({"인적사고1": test["인적사고"].map(temp_items).tolist()})
```

# Data Merge

- 제거한 컬럼

    - 필요없어 보여서: `발생일시` `사고인지 시간` `날씨` `기온` `습도` `연면적` `층 정보`

    - 결측치가 많아서: `물적사고`

- 추가한 컬럼

    - `Data Preprocessing` 과정을 적용한 컬럼 모두 추가

        - `공사종류` `공종` `사고객체` `장소` `부위` `인적사고`

```python
drop_cols = [
    "발생일시",
    "사고인지 시간",
    "날씨",
    "기온",
    "습도",
    "연면적",
    "층 정보",
    "물적사고",
]
```

```python
test_cp = test.copy()
test_cp = test_cp.drop(columns=drop_cols)

# merge
test_cp_concat = pd.concat([test_cp, a, b, c, d, e, f], axis=1)

# save
test_cp_concat.to_csv(f"{data_path}/{processed_filename}", index=False, encoding="utf-8-sig")
```
